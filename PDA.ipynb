Push Down Automata for a^n b^n

```python
class Stack:
    def __init__(self):
        self.l=[]
    def push(self,x):
        self.l.append(x)
        print(self.l)
    def pop(self):
        self.l.pop()
        print(self.l)
    def isempty(self):
        return self.l==[]
    def length(self):
        return len(self.l)
ip=Stack()
pda=dict()
pda[(0,'a')]=1
pda[(0,'b')]=4
pda[(0,'c')]=3
pda[(1,'a')]=1
pda[(1,'b')]=2
pda[(1,'c')]=4
pda[(2,'b')]=2
pda[(2,'a')]=4
pda[(2,'c')]=3
pda[(3,'c')]=3
pda[(3,'a')]=4
pda[(3,'b')]=4
pda[(4,'a')]=4
pda[(4,'b')]=4
pda[(4,'c')]=4
stack=[]
cs=0
fs=3
n=input("Enter a String :  ")
ip.push('c')
for i in n:
    if(i=='a'):
        ip.push(i)
        cs=pda[cs,i]
        print("Current State=",cs)
    elif(i=='b'):
        ip.pop()
        cs=pda[cs,i]
        print("Current State=",cs)
    if(i=='c'):
        cs=pda[cs,i]
        print("Current State=",cs)
if(ip.length()==1 and cs==fs):
    print("PDA is acceptable !!")
else:
    print("PDA cannot accept String !! ")
```

PDA IMPLEMENTATION OF EQUAL NUMBER OF a's and b's

```python
#test case 01
pda=dict()

pda[('q0','a')]='q0'
pda[('q0','b')]='q0'
pda[('q0','z')]='q1'
pda[('q2','a')]='q2'
pda[('q2','b')]='q2'
pda[('q2','z')]='q2'

class stack:
    def __init__ (self,pda):
        self.items=[]
    def push(self,item):
        self.items.append(item)
    def pop(self):
        return self.items.pop()
    def isempty(self):
        return self.items==[]
    def size(self):
        return len(self.items)
s=stack(pda)
x=input("Enter the string to be checked")
s.push('z')
cs='q0'
fs='q1'
for i in x:
    top=s.pop()
    s.push(top)
    cs=pda[cs,i]
    print("Current state is :",cs)
    if top=='z' and i=='a':
        s.push(i)
    elif top=='a' and i=='a':
        s.push(i)
    elif top=='z' and i=='b':
        s.push(i)
    elif top=='b' and i=='b':
        s.push(i)
    elif top=='a' and i=='b':
        s.pop()
    elif top=='b' and i=='a':
        s.pop()
    elif top!='a' and i=='b':
        s.push(i)
y=s.pop()
cs=pda[cs,y]
if cs==fs:
    print("String Accepted")
else:
    print("String Not Accepted")
```
